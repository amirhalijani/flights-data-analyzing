## Display all columns of the table along with the first 5 rows of data in the data frame

```python
import pandas as pd
df = pd.read_csv("flights.csv")
df.head()
```

## Display the last 10 rows of data in the dataframe

```python
df.tail(10)
```

## Mean Arrival Delay

```python
import matplotlib.pyplot as plt
from scipy import stats

trimmed_mean_arr_delay = stats.trim_mean(df['arr_delay'], 0.1)

plt.figure(figsize=(12, 6))

plt.plot(df.index, df['arr_delay'], color='blue', alpha=0.5, label='Arrival Delay')

plt.axhline(y=trimmed_mean_arr_delay, color='red', linestyle='--', label='Trimmed Mean Arrival Delay')

plt.text(0, trimmed_mean_arr_delay, f'Trimmed Mean: {trimmed_mean_arr_delay:.2f}', color='black')

plt.xlabel('Flight')
plt.ylabel('Delay')
plt.title('Arrival Delay Visualization with Trimmed Mean')
plt.legend()

plt.show()
```

## Mean Departure Delay

```python
import matplotlib.pyplot as plt
from scipy import stats

trimmed_mean_dep_delay = stats.trim_mean(df['dep_delay'], 0.1)

plt.figure(figsize=(12, 6))

plt.plot(df.index, df['dep_delay'], color='firebrick', alpha=0.5, label='Departure Delay')

plt.axhline(y=trimmed_mean_dep_delay, color='steelblue', linestyle='--', label='Trimmed Mean Departure Delay')

plt.text(0, trimmed_mean_arr_delay, f'Trimmed Mean: {trimmed_mean_dep_delay:.2f}', color='black')

plt.xlabel('Flight')
plt.ylabel('Delay')
plt.title('Departure Delay Visualization with Trimmed Mean')
plt.legend()

plt.show()
```

## A Visual Analysis of Flight Delays by Origin (Mean Delay)

```python
import matplotlib.pyplot as plt

origin_delay = df.groupby('origin')['dep_delay'].mean()

origin_delay_sorted = origin_delay.sort_values(ascending=False)

plt.figure(figsize=(45, 15))
origin_delay_sorted.plot(kind='bar', color='skyblue')
plt.xlabel('Origin')
plt.ylabel('Mean Delay')
plt.title('Mean Flight Delay by Origin')
plt.xticks(rotation=45)
plt.show()
```

## A Visual Analysis of Flight Delays by Destination (Mean Delay)

```python
import matplotlib.pyplot as plt

destination_delay = df.groupby('dest')['dep_delay'].mean()

destination_delay_sorted = destination_delay.sort_values(ascending=False)

plt.figure(figsize=(45, 15))
destination_delay_sorted.plot(kind='bar', color='mediumseagreen')
plt.xlabel('Destination')
plt.ylabel('Mean Delay')
plt.title('Mean Flight Delay by Destination')
plt.xticks(rotation=45)
plt.show()
```

## A Visual Analysis of Flight Delays by Destination (Total Delay)

```python
import matplotlib.pyplot as plt

destination_delay = df.groupby('dest')['dep_delay'].sum()

destination_delay_sorted = destination_delay.sort_values(ascending=False)

plt.figure(figsize=(45, 15))
destination_delay_sorted.plot(kind='bar', color='orange')
plt.xlabel('Destination')
plt.ylabel('Total Delay')
plt.title('Total Flight Delay by Destination')
plt.xticks(rotation=45)
plt.show()
```

## Arrivals Delay Histogram

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
df['arr_delay'].hist(bins=200, color='green', alpha=0.5)

plt.xlabel('Delay (minutes)')
plt.ylabel('Number of Flights')
plt.title('Histogram of Arrival Delays')
plt.grid(True)
plt.show()
```

## Delayed and on Time Flights Figure

```python
import matplotlib.pyplot as plt

total_flights = len(df)

delayed_flights = df[df['arr_delay'] > 10]['arr_delay'].count()

on_time_flights = df[df['arr_delay'] <= 10]['arr_delay'].count()

labels = ['Delayed Flights', 'On Time Flights']
sizes = [delayed_flights, on_time_flights]

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.2f%%', startangle=140, colors=['lightcoral', 'lightskyblue'])
plt.title('Percentage of Delayed and On Time Flights')
plt.axis('equal')
plt.show()
```

## Scatter Plot of Flight Time and Delay for Each Trip

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(df['dep_time'], df['dep_delay'], color='blue', alpha=0.5)

plt.xlabel('Departure Time')
plt.ylabel('Departure Delay (minutes)')
plt.title('Departure Delay vs. Departure Time')
plt.grid(True)
plt.show()
```

## Most Popular Destinations

```python
import matplotlib.pyplot as plt

destination_counts = df['dest'].value_counts()

top_destination = destination_counts.head(10)

plt.figure(figsize=(12, 8))
top_destination.plot(kind='bar', color='powderblue')
plt.xlabel('Destination')
plt.ylabel('Number of Flights')
plt.title('Top 10 Most Popular Destinations')
plt.xticks(rotation=45)
plt.show()
```

## Longest Average Departure Delay

```python
import matplotlib.pyplot as plt

average_delay_by_carrier = df.groupby('name')['dep_delay'].mean().nlargest(10)

plt.figure(figsize=(10, 6))
plt.bar(average_delay_by_carrier.index, average_delay_by_carrier.values, color='skyblue')
plt.xlabel('Airline Carrier')
plt.ylabel('Average Departure Delay')
plt.title('Top 10 Airline Carriers with the Longest Average Departure Delays')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Shortest Average Departure Delay

```python
import matplotlib.pyplot as plt

average_delay_by_carrier = df.groupby('name')['dep_delay'].mean().nsmallest(5)

plt.figure(figsize=(10, 6))
plt.bar(average_delay_by_carrier.index, average_delay_by_carrier.values, color='deepskyblue')
plt.xlabel('Airline Carrier')
plt.ylabel('Average Departure Delay')
plt.title('Top 5 Airline Carriers with the Shortest Average Departure Delays')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
